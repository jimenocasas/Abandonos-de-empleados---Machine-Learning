# Modelo final

Este cuaderno se centra en el entrenamiento de un modelo de clasificaci贸n SVM con kernel RBF (Radial Basis Function) utilizando hiperpar谩metros 贸ptimos. Primero, el modelo se entrena con los datos procesados y se eval煤a utilizando m茅tricas como Balanced Accuracy, Sensibilidad (TPR) y Especificidad (TNR), adem谩s de generar un reporte de clasificaci贸n. La matriz de confusi贸n y las m茅tricas de evaluaci贸n se visualizan para facilitar la interpretaci贸n. Posteriormente, el modelo entrenado se guarda para su uso posterior. Finalmente, se realiza una predicci贸n sobre un conjunto de datos de la competencia y se almacenan los resultados en un archivo CSV. Se incluye un gr谩fico para visualizar la distribuci贸n de las predicciones.

```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
```

##  Entrenar el Modelo Definitivo (SVM RBF con Hiperpar谩metros ptimos)

En esta secci贸n, entrenamos el modelo **SVM con kernel RBF** utilizando los hiperpar谩metros 贸ptimos. Los datos de entrenamiento son cargados y procesados antes de ser utilizados para entrenar el modelo.


```python
# Entrenar el modelo definitivo (SVM RBF con hiperpar谩metros 贸ptimos)
modelo_final = SVC(kernel='rbf', class_weight='balanced', C=5, gamma=0.1)

# Entrenar el modelo con los datos procesados
X_train = pd.read_csv("../results/X_train.csv")
y_train = pd.read_csv("../results/y_train.csv").values.ravel()
modelo_final.fit(X_train, y_train)
```

##  Evaluaci贸n del Modelo

Aqu铆 realizamos las predicciones con el modelo entrenado, comparamos los resultados con los valores reales, y calculamos m茅tricas como **Balanced Accuracy**, **Sensibilidad (TPR)** y **Especificidad (TNR)**.


```python
# Cargamos el predictor del modelo final
X_test = pd.read_csv("../results/X_test.csv")  # El X_test ya est谩 procesado

y_predict = modelo_final.predict(X_test)

# Hacemos la comparaci贸n de las predicciones con los valores reales
y_test = pd.read_csv("../results/y_test.csv").values.ravel()
balanced_accuracy = balanced_accuracy_score(y_test, y_predict)
matriz_conf = confusion_matrix(y_test, y_predict)

# Calcular Sensibilidad (TPR) y Especificidad (TNR)
tn, fp, fn, tp = matriz_conf.ravel()
tpr = tp / (tp + fn)  # Sensibilidad
tnr = tn / (tn + fp)  # Especificidad

# Imprimir m茅tricas
print(f"\n Balanced Accuracy: {balanced_accuracy:.4f}")
print(f"锔 Sensibilidad (TPR): {tpr:.4f}")
print(f"锔 Especificidad (TNR): {tnr:.4f}")
print("\n Reporte de clasificaci贸n:\n", classification_report(y_test, y_predict))
```

##  Visualizaci贸n de M茅tricas

Aqu铆 visualizamos las m茅tricas de evaluaci贸n del modelo, como **Balanced Accuracy**, **Sensibilidad (TPR)** y **Especificidad (TNR)**. Tambi茅n mostramos la **matriz de confusi贸n** en un gr谩fico.


```python
# Visualizaci贸n de m茅tricas
plt.figure(figsize=(12, 5))

# Matriz de confusi贸n
plt.subplot(1, 2, 1)
sns.heatmap(matriz_conf, annot=True, fmt='d', cmap='Blues', xticklabels=["No", "Yes"], yticklabels=["No", "Yes"])
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusi贸n")

# Balanced Accuracy, TPR y TNR
plt.subplot(1, 2, 2)
metrics = [balanced_accuracy, tpr, tnr]
names = ["Balanced Accuracy", "TPR (Sensibilidad)", "TNR (Especificidad)"]
plt.barh(names, metrics, color=["blue", "green", "red"])
plt.xlim(0, 1)
plt.xlabel("Valor")
plt.title("M茅tricas de Evaluaci贸n del Modelo")

plt.tight_layout()
plt.show()
```

##  Guardar el Modelo Entrenado

Guardamos el modelo entrenado en un archivo **`modelo_final.pkl`** para su posterior uso.


```python
# Guardar el modelo entrenado
df = pd.read_csv("../attrition_datasets/train_test/attrition_availabledata_07"
                 ".csv/attrition_availabledata_07.csv")
preprocesado = joblib.load("../results/processed_data.pkl")

X = df.drop(columns=["Attrition", "EmployeeID", "EmployeeCount", "Over18", "StandardHours"], errors='ignore')  # Eliminamos la columna objetivo

X_procesado = preprocesado.transform(X)

y = df["Attrition"].map({"Yes": 1, "No": 0})  # Convertimos 'Yes' y 'No' a valores num茅ricos

mejor_modelo = modelo_final.fit(X_procesado, y)

joblib.dump(mejor_modelo, "../results/modelo_final.pkl")

print("\n 隆Entrenamiento completado! El modelo ha sido guardado como 'modelo_final.pkl'.")
```

##  Cargar los Datos de la Competici贸n

Aqu铆 cargamos los datos de la competencia (sin la variable objetivo) y aplicamos la misma transformaci贸n que en los datos de entrenamiento.


```python
# Cargar los datos de la competici贸n (sin la variable objetivo)
df_comp = pd.read_csv("../attrition_datasets/train_test/attrition_competition_07.csv/attrition_competition_07.csv")

# Preprocesar los datos de la competici贸n antes de predecir
X_comp = df_comp.drop(columns=["EmployeeCount", "Over18", "StandardHours"], errors='ignore')
X_comp_procesado = preprocesado.transform(X_comp)  # Aplicamos la transformaci贸n
```

##  Generar Predicciones con el Modelo

Generamos las predicciones para los datos de la competencia utilizando el modelo entrenado previamente.


```python
# Generar predicciones con el modelo ya entrenado
predicciones_competicion = modelo_final.predict(X_comp_procesado)
```

##  Guardar las Predicciones

Guardamos las predicciones generadas en un archivo CSV para su posterior evaluaci贸n en la competici贸n.


```python
# Guardar las predicciones en un archivo CSV
predicciones_df = pd.DataFrame({
    "EmployeeID": df_comp["EmployeeID"],
    "Attrition": pd.Series(predicciones_competicion).map({1: "Yes", 0: "No"})
})
predicciones_df.to_csv("../results/predicciones.csv", index=False)
print("\n 隆Predicciones generadas y guardadas en 'predicciones.csv'!")
```

##  Crear un Gr谩fico de Predicciones

Finalmente, creamos un gr谩fico de las predicciones generadas para visualizar la distribuci贸n de las predicciones.


```python
# Hacemos un conteo de las predicciones y lo mostramos en una gr谩fica
print("\n  Creando gr谩fico de predicciones...")
plt.figure(figsize=(6, 4))
ax = sns.countplot(x="Attrition", hue="Attrition", data=predicciones_df, palette="viridis", legend=False)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')
plt.title("Predicciones de Attrition en Competici贸n")
plt.show()
```
